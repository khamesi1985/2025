```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

# بارگذاری داده‌ها
url = "https://raw.githubusercontent.com/khamesi1985/2025/main/wdbc.data"
data = pd.read_csv(url, header=None)
data = data.dropna()
X = data.iloc[:, 2:32]
Pre_Y = data.iloc[:, 1]
make_bin = LabelEncoder()
make_bin.fit(Pre_Y)
Y = make_bin.transform(Pre_Y)

# تقسیم داده‌ها به آموزش و آزمون
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# استاندارد سازی داده ها
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# پیاده سازی مدل رگرسیون لاجستیک
LGR_model = LogisticRegression(random_state=42)
LGR_scores = cross_val_score(LGR_model, X_train_scaled, Y_train, cv=4, scoring='accuracy')
LGR_model.fit(X_train_scaled, Y_train)
Y_pred_train_LGR = LGR_model.predict(X_train_scaled)
Y_pred_test_LGR = LGR_model.predict(X_test_scaled)

# محاسبه دقت مدل رگرسیون لاجستیک
accuracy_test_LGR = accuracy_score(Y_test, Y_pred_test_LGR)
precision_test_LGR = precision_score(Y_test, Y_pred_test_LGR)
recall_test_LGR = recall_score(Y_test, Y_pred_test_LGR)
f1_score_test_LGR = f1_score(Y_test, Y_pred_test_LGR)
print("LOGISTIC REGRESSION-LGR CLASSIFIER")
print("Train Data Evaluate via Cross Validation by LGR :")
print(f"Accuracy Scores for Each Fold By LGR : {LGR_scores}")
print(f"Mean Accuracy score by LGR : {np.mean(LGR_scores):.6f}")
print(classification_report(Y_train, Y_pred_train_LGR, target_names=make_bin.classes_))
print("Test Data Evaluate by LGR :")
print("Accuracy On Test Data by LGR = ", accuracy_test_LGR)
print("Precision On Test Data by LGR = ", precision_test_LGR)
print("Recall On Test Data by LGR = ", recall_test_LGR)
print("F1-Score On Test Data by LGR = ", f1_score_test_LGR)
len_test = len(Y_test)
TN = TP = FN = FP = 0
for i in range (len_test):
  if Y_test[i] == 0 and Y_pred_test_LGR[i] == 1:
    FP = FP + 1
  elif Y_test[i] == 1 and Y_pred_test_LGR[i] == 0:
    FN = FN + 1
  elif Y_test[i] == 1 and Y_pred_test_LGR[i] == 1:
    TP = TP + 1
  elif Y_test[i] == 0 and Y_pred_test_LGR[i] == 0:
    TN = TN + 1
print("TP On Test Data by LGR = ", TP)
print("TN On Test Data by LGR = ", TN)
print("FP On Test Data by LGR = ", FP)
print("FN On Test Data by LGR = ", FN)
print(classification_report(Y_test, Y_pred_test_LGR, target_names=make_bin.classes_))
print("-"*55)

# پیاده سازی مدل ماشین بردار پشتیبان
SVM_model = SVC()
SVM_scores = cross_val_score(SVM_model, X_train_scaled, Y_train, cv=4, scoring='accuracy')
SVM_model.fit(X_train_scaled, Y_train)
Y_pred_train_SVM = SVM_model.predict(X_train_scaled)
Y_pred_test_SVM = SVM_model.predict(X_test_scaled)

# محاسبه دقت مدل ماشین بردار پشتیبان
accuracy_test_SVM = accuracy_score(Y_test, Y_pred_test_SVM)
precision_test_SVM = precision_score(Y_test, Y_pred_test_SVM)
recall_test_SVM = recall_score(Y_test, Y_pred_test_SVM)
f1_score_test_SVM = f1_score(Y_test, Y_pred_test_SVM)
print("SUPPORT VECTOR MACHINE-SVM CLASSIFIER")
print("Train Data Evaluate via Cross Validation by SVM :")
print(f"Accuracy Scores for Each Fold By SVM : {SVM_scores}")
print(f"Mean Accuracy score by SVM : {np.mean(SVM_scores):.6f}")
print(classification_report(Y_train, Y_pred_train_SVM, target_names=make_bin.classes_))
print("Test Data Evaluate by SVM :")
print("Accuracy On Test Data by SVM = ", accuracy_test_SVM)
print("Precision On Test Data by SVM = ", precision_test_SVM)
print("Recall On Test Data by SVM = ", recall_test_SVM)
print("F1-Score On Test Data by SVM = ", f1_score_test_SVM)
len_test = len(Y_test)
TN = TP = FN = FP = 0
for i in range (len_test):
  if Y_test[i] == 0 and Y_pred_test_SVM[i] == 1:
    FP = FP + 1
  elif Y_test[i] == 1 and Y_pred_test_SVM[i] == 0:
    FN = FN + 1
  elif Y_test[i] == 1 and Y_pred_test_SVM[i] == 1:
    TP = TP + 1
  elif Y_test[i] == 0 and Y_pred_test_SVM[i] == 0:
    TN = TN + 1
print("TP On Test Data by SVM = ", TP)
print("TN On Test Data by SVM = ", TN)
print("FP On Test Data by SVM = ", FP)
print("FN On Test Data by SVM = ", FN)
print(classification_report(Y_test, Y_pred_test_SVM, target_names=make_bin.classes_))
print("-"*55)

# پیاده سازی مدل (k) نزدیک ترین همسایه
KNN_model = KNeighborsClassifier()
KNN_scores = cross_val_score(KNN_model, X_train_scaled, Y_train, cv=4, scoring='accuracy')
KNN_model.fit(X_train_scaled, Y_train)
Y_pred_train_KNN = KNN_model.predict(X_train_scaled)
Y_pred_test_KNN = KNN_model.predict(X_test_scaled)

# محاسبه دقت مدل (k) نزدیک ترین همسایه
accuracy_test_KNN = accuracy_score(Y_test, Y_pred_test_KNN)
precision_test_KNN = precision_score(Y_test, Y_pred_test_KNN)
recall_test_KNN = recall_score(Y_test, Y_pred_test_KNN)
f1_score_test_KNN = f1_score(Y_test, Y_pred_test_KNN)
print("K NEAREST NEIGHBORS-KNN CLASSIFIER")
print("Train Data Evaluate via Cross Validation by KNN :")
print(f"Accuracy Scores for Each Fold By KNN : {KNN_scores}")
print(f"Mean Accuracy score by KNN : {np.mean(KNN_scores):.6f}")
print(classification_report(Y_train, Y_pred_train_KNN, target_names=make_bin.classes_))
print("Test Data Evaluate by KNN :")
print("Accuracy On Test Data by KNN = ", accuracy_test_KNN)
print("Precision On Test Data by KNN = ", precision_test_KNN)
print("Recall On Test Data by KNN = ", recall_test_KNN)
print("F1-Score On Test Data by KNN = ", f1_score_test_KNN)
len_test = len(Y_test)
TN = TP = FN = FP = 0
for i in range (len_test):
  if Y_test[i] == 0 and Y_pred_test_KNN[i] == 1:
    FP = FP + 1
  elif Y_test[i] == 1 and Y_pred_test_KNN[i] == 0:
    FN = FN + 1
  elif Y_test[i] == 1 and Y_pred_test_KNN[i] == 1:
    TP = TP + 1
  elif Y_test[i] == 0 and Y_pred_test_KNN[i] == 0:
    TN = TN + 1
print("TP On Test Data by KNN = ", TP)
print("TN On Test Data by KNN = ", TN)
print("FP On Test Data by KNN = ", FP)
print("FN On Test Data by KNN = ", FN)
print(classification_report(Y_test, Y_pred_test_KNN, target_names=make_bin.classes_))
print("-"*55)

# پیاده سازی مدل درخت تصمیم
DT_model = DecisionTreeClassifier(random_state=42, class_weight='balanced')
DT_scores = cross_val_score(DT_model, X_train_scaled, Y_train, cv=4, scoring='accuracy')
DT_model.fit(X_train_scaled, Y_train)
Y_pred_train_DT = DT_model.predict(X_train_scaled)
Y_pred_test_DT = DT_model.predict(X_test_scaled)

# محاسبه دقت مدل درخت تصمیم
accuracy_test_DT = accuracy_score(Y_test, Y_pred_test_DT)
precision_test_DT = precision_score(Y_test, Y_pred_test_DT)
recall_test_DT = recall_score(Y_test, Y_pred_test_DT)
f1_score_test_DT = f1_score(Y_test, Y_pred_test_DT)
print("DECISION TREE-DT CLASSIFIER")
print("Train Data Evaluate via Cross Validation by DT :")
print(f"Accuracy Scores for Each Fold By DT : {DT_scores}")
print(f"Mean Accuracy score by DT : {np.mean(DT_scores):.6f}")
print(classification_report(Y_train, Y_pred_train_DT, target_names=make_bin.classes_))
print("Test Data Evaluate by DT :")
print("Accuracy On Test Data by DT = ", accuracy_test_DT)
print("Precision On Test Data by DT = ", precision_test_DT)
print("Recall On Test Data by DT = ", recall_test_DT)
print("F1-Score On Test Data by DT = ", f1_score_test_DT)
len_test = len(Y_test)
TN = TP = FN = FP = 0
for i in range (len_test):
  if Y_test[i] == 0 and Y_pred_test_DT[i] == 1:
    FP = FP + 1
  elif Y_test[i] == 1 and Y_pred_test_DT[i] == 0:
    FN = FN + 1
  elif Y_test[i] == 1 and Y_pred_test_DT[i] == 1:
    TP = TP + 1
  elif Y_test[i] == 0 and Y_pred_test_DT[i] == 0:
    TN = TN + 1
print("TP On Test Data by DT = ", TP)
print("TN On Test Data by DT = ", TN)
print("FP On Test Data by DT = ", FP)
print("FN On Test Data by DT = ", FN)
print(classification_report(Y_test, Y_pred_test_DT, target_names=make_bin.classes_))
print("-"*55)

# پیاده سازی مدل جنگل تصادفی
RF_model = RandomForestClassifier(random_state=42)
RF_scores = cross_val_score(RF_model, X_train_scaled, Y_train, cv=4, scoring='accuracy')
RF_model.fit(X_train_scaled, Y_train)
Y_pred_train_RF = RF_model.predict(X_train_scaled)
Y_pred_test_RF = RF_model.predict(X_test_scaled)

# محاسبه دقت مدل جنگل تصادفی
accuracy_test_RF = accuracy_score(Y_test, Y_pred_test_RF)
precision_test_RF = precision_score(Y_test, Y_pred_test_RF)
recall_test_RF = recall_score(Y_test, Y_pred_test_RF)
f1_score_test_RF = f1_score(Y_test, Y_pred_test_RF)
print("RANDOM FOREST-RF CLASSIFIER")
print("Train Data Evaluate via Cross Validation by RF :")
print(f"Accuracy Scores for Each Fold By RF : {RF_scores}")
print(f"Mean Accuracy score by RF : {np.mean(RF_scores):.6f}")
print(classification_report(Y_train, Y_pred_train_RF, target_names=make_bin.classes_))
print("Test Data Evaluate by RF :")
print("Accuracy On Test Data by RF = ", accuracy_test_RF)
print("Precision On Test Data by RF = ", precision_test_RF)
print("Recall On Test Data by RF = ", recall_test_RF)
print("F1-Score On Test Data by RF = ", f1_score_test_RF)
len_test = len(Y_test)
TN = TP = FN = FP = 0
for i in range (len_test):
  if Y_test[i] == 0 and Y_pred_test_RF[i] == 1:
    FP = FP + 1
  elif Y_test[i] == 1 and Y_pred_test_RF[i] == 0:
    FN = FN + 1
  elif Y_test[i] == 1 and Y_pred_test_RF[i] == 1:
    TP = TP + 1
  elif Y_test[i] == 0 and Y_pred_test_RF[i] == 0:
    TN = TN + 1
print("TP On Test Data by RF = ", TP)
print("TN On Test Data by RF = ", TN)
print("FP On Test Data by RF = ", FP)
print("FN On Test Data by RF = ", FN)
print(classification_report(Y_test, Y_pred_test_RF, target_names=make_bin.classes_))
print("-"*55)

# پیاده سازی مدل شبکه عصبی
ANN_model = MLPClassifier(max_iter=700, random_state=42)
ANN_scores = cross_val_score(ANN_model, X_train_scaled, Y_train, cv=4, scoring='accuracy')
ANN_model.fit(X_train_scaled, Y_train)
Y_pred_train_ANN = ANN_model.predict(X_train_scaled)
Y_pred_test_ANN = ANN_model.predict(X_test_scaled)

# محاسبه دقت مدل شبکه عصبی
accuracy_test_ANN = accuracy_score(Y_test, Y_pred_test_ANN)
precision_test_ANN = precision_score(Y_test, Y_pred_test_ANN)
recall_test_ANN = recall_score(Y_test, Y_pred_test_ANN)
f1_score_test_ANN = f1_score(Y_test, Y_pred_test_ANN)
print("ARTIFICIAL NEURAL NETWORK-ANN CLASSIFIER")
print("Train Data Evaluate via Cross Validation by ANN :")
print(f"Accuracy Scores for Each Fold By ANN : {ANN_scores}")
print(f"Mean Accuracy score by ANN : {np.mean(ANN_scores):.6f}")
print(classification_report(Y_train, Y_pred_train_ANN, target_names=make_bin.classes_))
print("Test Data Evaluate by ANN :")
print("Accuracy On Test Data by ANN = ", accuracy_test_ANN)
print("Precision On Test Data by ANN = ", precision_test_ANN)
print("Recall On Test Data by ANN = ", recall_test_ANN)
print("F1-Score On Test Data by ANN = ", f1_score_test_ANN)
len_test = len(Y_test)
TN = TP = FN = FP = 0
for i in range (len_test):
  if Y_test[i] == 0 and Y_pred_test_ANN[i] == 1:
    FP = FP + 1
  elif Y_test[i] == 1 and Y_pred_test_ANN[i] == 0:
    FN = FN + 1
  elif Y_test[i] == 1 and Y_pred_test_ANN[i] == 1:
    TP = TP + 1
  elif Y_test[i] == 0 and Y_pred_test_ANN[i] == 0:
    TN = TN + 1
print("TP On Test Data by ANN = ", TP)
print("TN On Test Data by ANN = ", TN)
print("FP On Test Data by ANN = ", FP)
print("FN On Test Data by ANN = ", FN)
print(classification_report(Y_test, Y_pred_test_ANN, target_names=make_bin.classes_))
print("-"*55)
```
